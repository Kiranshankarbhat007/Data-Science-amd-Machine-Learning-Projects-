### Feature Engineering 

* We will perform some Feature engineering Steps for
    * Missing Values
    * Outliars
    * Categorical feature conversion
    * Temporal variables

#### For Training Data

#### For Missing Values

```python
train_cat_null = [x for x in train_data.columns if train_data[x].isnull().sum()>1 and train_data[x].dtype == "O"]

for x in train_cat_null:
    print ('{} has {} % of missing value'.format(x , np.round(train_data[x].isnull().mean()*100,3)))
    
train_cat_null
```

```python
# Replacing the missing Values by a new categorical value "Nan"

def replace_cat (dataset, feature):
    data = train_data.copy()
    data[train_cat_null] = data[train_cat_null].fillna('Missing')
    return data

train_data = replace_cat(train_data, train_cat_null)

train_data[train_cat_null].isnull().sum()
```

```python
train_data.head()
```

```python
# Now for numerical feature 

train_num_null = [x for x in train_data.columns if train_data[x].isnull().sum()>1 and train_data[x].dtype!='O']

for x in train_num_null:
    print ('{}: {}% of Null values'.format(x, np.round(train_data[x].isnull().mean()*100,3)))
```

```python
for x in train_num_null:
    median_value = train_data[x].median()
    
    train_data[x+'_nan'] = np.where(train_data[x].isnull(),1,0)
    train_data[x].fillna(median_value,inplace=True)
    
train_data[train_num_null].isnull().sum()
```

```python
train_data.head()
```

```python
# Temporal Variables (Date Time Variables)

for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
       
    train_data[feature]=train_data['YrSold']-train_data[feature]
```

```python
train_data[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

### Numerical Features

```python
train_data.head()
```

```python
num_features=['LotFrontage', 'LotArea', '1stFlrSF', 'GrLivArea', 'SalePrice']

for x in num_features:
    train_data[x] = np.log(train_data[x])
```

```python
train_data.head()
```

### Handling Rare Categorical Feature
#### We will remove categorical variables that are present less than 1% of the observations

```python
train_cat_feature
```

```python
for feature in train_cat_feature:
    temp= train_data.groupby(feature)['SalePrice'].count()/len(train_data)
    temp_df= temp[temp>0.01].index
    train_data[feature]= np.where(train_data[feature].isin(temp_df),train_data[feature],'Rare_var')
```

```python
train_data.head(10)
```

```python
for feature in train_cat_feature:
    labels_ordered=train_data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    train_data[feature]=train_data[feature].map(labels_ordered)
```

```python
train_data.head()
```

### Feature Scaling 

```python
feature_scale =[feature for feature in train_data.columns if feature not in ['Id','SalePrice']]

from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(train_data[feature_scale])
```

```python
scaler.transform(train_data[feature_scale])
```

```python
# transform the train and add on the Id and SalePrice variables
data_train = pd.concat([train_data[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(train_data[feature_scale]), columns=feature_scale)],
                    axis=1)
```

```python
data_train.to_csv('Final_train.csv',index=False)
```

### Feature selection 

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
x = data.drop(['Id','SalePrice'],axis=1)
y = data['SalePrice']
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(x,y)
```

```python
feature_sel_model.get_support()
```

```python
selected_feat = x.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((x.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
print('features with coefficients shrank to zero: {}'.format(
    np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
selected_feat
```

```python
x = x[selected_feat]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.2)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
result = pd.DataFrame({"SalePrice_real": y_test, "SalePrice PREDICTION": y_pred})
result
```
